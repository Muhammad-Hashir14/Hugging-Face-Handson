```python
import pandas as pd
from transformers import AutoTokenizer

def tokenization(batch):
    # Check if input is a DataFrame
    if not isinstance(batch, pd.DataFrame):
        raise ValueError("Input must be a pandas DataFrame.")

    # Extract 'text' column and convert to list
    texts = batch['text'].tolist()

    # Initialize tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

    # Perform tokenization
    tokenized = tokenizer(texts, padding=True, truncation=True, return_tensors=None)

    # Extract input_ids and attention_mask
    input_ids = tokenized['input_ids']
    attention_mask = tokenized['attention_mask']

    # Create new columns for tokenized data
    batch['input_ids'] = [ids for ids in input_ids]
    batch['attention_mask'] = [mask for mask in attention_mask]

    return batch
```

##Evaluation Metrics of LLMs

###Bert Score



```python
# Install the bert_score library
!pip install bert_score

# Import the necessary functions
from bert_score import score

# Prepare your predictions and references
predictions = ["hello world", "general kenobi"]
references = ["hello world", "general kenobi"]

# Compute BERTScore specifying the language (English)
precision, recall, f1 = score(predictions, references, lang="en")

# Print the results
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

###Bleu Score

```python
import nltk
from nltk.translate.bleu_score import sentence_bleu

# Define candidate and reference sentences
candidate = "hello world"
reference = ["hello world", "general kenobi"]

# Tokenize candidate and reference sentences
candidate_tokens = candidate.split()
reference_tokens = [ref.split() for ref in reference]

# Calculate BLEU score
bleu_score = sentence_bleu(reference_tokens, candidate_tokens)

# Print BLEU score
print("BLEU Score:", bleu_score)
```
